## STEP 1 - Reading from file

## STEP 2 - Data cleansing

```python
# Read the file char by char (up to 4 bytes only),
# convert all letters to lowercase,
# remove all non-letter characters,
# and replace consecutive separators with a single space
#
# The output will be saved into a file
#
# The output will be just a single line of lowercase words
# separated by a single space
```

```python
word = ''
word_count = 0
char_count = 0
separator = False
file_out = open('output.txt', 'w')

with open('text2.txt', 'r') as file_in:
    ch = file_in.read(1)
    while (ch):
        # increment char count
        char_count = char_count + 1
        # convert char to lowercase
        ch = ch.lower()
        # print(ch, end='')
        
        if (ch in letters):
            word = word + ch
            separator = True
        else:
            if (separator):
                # print(word)
                file_out.write(word + ' ')
                word_count = word_count + 1
                word = ''
                separator = False
                
        # read the next char from the file
        ch = file_in.read(1)
#
file_out.close()
print('Done! Words found:', word_count, 'and there are', char_count, 'chars')
```

## STEP 3 - Get words

```python
# Open cleaned file and handle each word
# Remember, the input file contains just a single line of text (no matter how huge it is),
# that's why it's not possible to read this file line by line - it must be loaded char by char
```

```python
this_is_a_word = False
word = ''
word_count = 0
cloud = {}

with open('output.txt', 'r') as file_in:
    ch = file_in.read(1) # read one character
    while (ch):
        # print(ch, end='')
        if (ch != ' '):
            this_is_a_word = True
            word = word + ch
        else:            
            if (this_is_a_word):
                # print(word, end='\n')  
                
                # check if word is already in the dictionary
                if (word in cloud): 
                    # if yes - increment the counter
                    cloud[word] = cloud[word] + 1
                else:                
                    # if not - it's a new word, let's add it to the dictionary
                    cloud[word] = 1               
                #
                # prepare for the next word
                #
                # increment word counter
                word_count = word_count + 1      
                # reset the flag
                this_is_a_word = False                
                # reset the word
                word = ''                
                
        # read the next character
        ch = file_in.read(1)
file_in.close() 
print('Done! Words found:', word_count)
print('Cloud size:', len(cloud))
```

```python
print(cloud)
```

## Post-processing 1: Filter stop words

```python
# apply filter to remove common words that don't carry any information (e.g. the a of)
```

```python
print('Cloud size before the filter:', len(cloud))
stoplist = ['and', 'of', 'to', 'the', 'that', 'it', 'is', 'in', 'for',
            'these', 'by', 'such', 'as', 'are', 'a', 'this', 'or', 'can',
            'be', 'which', 'their', 'an', 'on', 'from', 's', 'has', 'into',
            'they', 'also', 'with']
for i in stoplist:
    if (i in cloud):
        del cloud[i]
print('Cloud size after  the filter:', len(cloud))
```

## Post-processing 2: Delete items below certain threshold

```python
cutoff = 9 # any word count BELOW (strictly less) this number will be removed
items = []
for i,v in cloud.items():
    #print('The key is {} and the value is {}'.format(i, v))
    if (v < cutoff):
        items.append(i)
for i in items:
    del cloud[i]
print('Cloud size:', len(cloud))
```

## Post-processing 3: Sorting

```python
# it's not possible to sort a dictionary
# however, it's possible to sort it's representation (copy)
```

```python
import operator
# sorted_by_key = sorted(cloud.items(), key=operator.itemgetter(0))
sorted_by_value = sorted(cloud.items(), key=operator.itemgetter(1), reverse=True)
sorted_by_value
```

```python
# alternative sorting
sorted_by_value2 = sorted(cloud.items(), key=lambda x: x[1], reverse=True)
sorted_by_value2
```

## Post-processing 4: Get top N words

```python
top = 10
word_cloud = sorted_by_value[0:top]
word_cloud
```
